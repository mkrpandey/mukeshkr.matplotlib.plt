```python
#and also we can change the width size as per need
#and also can change the colour of the plot
#and also can change the diffrent-2 colour of the plot
#and also can change the colour of the plot boundry, linewidth
#and also can add label with the help of legend 


import matplotlib.pyplot as plt
x=["Python","Java","HTML","C",]
y=[8,5,6,7.8]

plt.xlabel("Language",fontsize=15)
plt.ylabel("No. of popularity",fontsize=15)
plt.title("Language Popularity",fontsize=18)
c=["blue","brown","green","red"]
plt.bar(x,y,width=0.5,color=c,edgecolor="black",linewidth=5,label="LangPop")
plt.legend()
plt.show()
```

```python
#Now apply all function apply on same colour

import matplotlib.pyplot as plt
x=["Python","Java","HTML","C",]
y=[8,5,6,7.8]

plt.xlabel("Language",fontsize=15)
plt.ylabel("No. of popularity",fontsize=15)
plt.title("Language Popularity",fontsize=18)
plt.bar(x,y,width=0.5,color="skyblue",edgecolor="black",linewidth=4,label="LangPop")
plt.legend()
plt.show()
```

```python
#Double Bar graph in same fig with overlap


import matplotlib.pyplot as plt
x=["Python","Java","HTML","C",]
y=[20,12,16,19]
z=[11,6,9,12]
plt.xlabel("Language",fontsize=15)
plt.ylabel("No. of popularity",fontsize=15)
plt.title("Language Popularity",fontsize=18)
plt.bar(x,y,width=0.5,color="skyblue",edgecolor="black",label="LangPop1")
plt.bar(x,z,width=0.5,color="lightblue",edgecolor="black",label="LangPop2")
plt.legend()
plt.show()
```

```python
#Double Bar graph in same fig without overlap


import matplotlib.pyplot as plt
import numpy as np

x=["Python","Java","HTML","C",]
y=[20,12,16,19]
z=[11,6,9,12]

width=0.4
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Language",fontsize=15)
plt.ylabel("No. of popularity",fontsize=15)
plt.title("Language Popularity",fontsize=18)


plt.bar(p,y,width,color="skyblue",edgecolor="black",label="LangPop1")
plt.bar(p1,z,width,color="lightblue",edgecolor="black",label="LangPop2")

plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
#Create a vertical bar graph


import matplotlib.pyplot as plt

x=["Python","Java","HTML","C",]
y=[20,12,16,19]
z=[11,6,9,12]

plt.xlabel("No. of popularity",fontsize=15)
plt.ylabel("Language",fontsize=15)
plt.title("Language Popularity",fontsize=18)

plt.barh(x,y,width,color="skyblue",edgecolor="black",label="LangPop1")
plt.barh(x,z,width,color="lightblue",edgecolor="black",label="LangPop2")

plt.legend()
plt.show()
```

# #Scatter Plot

```python
#simple scatter plot program with same color

import matplotlib.pyplot as plt

day=[1,2,3,4,5,6,7]
person=[2,3,1,4,5,3,6]

plt.scatter(day,person,color="r")

plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=12)
plt.ylabel("Person",fontsize=12)

plt.show
```

```python
#simple scatter plot program with diff-2 color

import matplotlib.pyplot as plt

day=[1,2,3,4,5,6,7]
person=[2,3,1,4,5,3,6]

colors=["red","blue","green","black","darkgreen","brown","red"]
plt.scatter(day,person,c=colors)

plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=12)
plt.ylabel("Person",fontsize=12)

plt.show
```

```python
#we can change(increase,decrease) the size of scatter plot dot(point)
#for color transparancy use (Ex- alpha=0.5)
import matplotlib.pyplot as plt

day=[1,2,3,4,5,6,7]
person=[2,3,1,4,5,3,6]

colors=["red","blue","green","black","darkgreen","brown","red"]
size=[200]
plt.scatter(day,person,c=colors,s=size)

plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=12)
plt.ylabel("Person",fontsize=12)

plt.show
```

```python
#we can change(increase,decrease) the size of scatter plot dot(point)
#we can change the shape of scatter plot use marker parameter (Ex-marker="*")
#for color transparancy use alpha parameter (Ex- alpha=0.5)


import matplotlib.pyplot as plt

day=[1,2,3,4,5,6,7]
person=[2,3,1,4,5,3,6]

colors=["red","blue","green","black","darkgreen","brown","red"]
size=[200]
plt.scatter(day,person,c=colors,s=size,marker="*")

plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=12)
plt.ylabel("Person",fontsize=12)

plt.show
```

# #Stack Plot

```python
#Simple Stackplot(Area) print

import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area=[2,3,2,5,4]

plt.stackplot(x,area)

plt.show()
```

```python
#Multiple Stackplot(Area) in one fig. print

import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area=[2,3,2,5,4] 
area1=[2,3,4,5,6]
area2=[1,3,2,4,2]

plt.stackplot(x,area,area1,area2)

plt.show()
```

```python
#Multiple Stackplot(Area) in one fig. print
#Modify in stackplot--- Add Label with [plt.legend.()]


import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area=[2,3,2,5,4] 
area1=[2,3,4,5,6]
area2=[1,3,2,4,2]

l=["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l)

plt.legend(loc=2)
plt.show()
```

```python
#Multiple Stackplot(Area) in one fig. print
#Modify in stackplot--- Can change the area plot color


import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area=[2,3,2,5,4] 
area1=[2,3,4,5,6]
area2=[1,3,2,4,2]

l=["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l,colors=["skyblue","darkturquoise","brown"])

plt.legend(loc=2)
plt.show()
```

```python
#Multiple Stackplot(Area) in one fig. print
#Modify in stackplot--- Add Title, xlabel, ylabel


import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area=[2,3,2,5,4] 
area1=[2,3,4,5,6]
area2=[1,3,2,4,2]

l=["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l,colors=["skyblue","darkturquoise","brown"])

plt.title("Python",fontsize=20)
plt.xlabel("x-axis",fontsize=15)
plt.ylabel("y-axis",fontsize=15)
plt.legend(loc=2)
plt.show()
```

```python
#Simple print
#Add color, Title, Xlabel, Ylabel
#use fill_between parameter


import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color="r")

plt.fill_between(x=[2,4],y1=2,y2=4)

plt.title("Pyhton")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
